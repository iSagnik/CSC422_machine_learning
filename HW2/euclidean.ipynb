```python
from sklearn.metrics.pairwise import euclidean_distances

X = [
    [5.56, 1.25],
    [3.61, 3.33],
    [8.06, 5.0],
    [3.89, 4.17],
    [10.0, 7.5],
    [2.78, 7.08],
    [1.94, 0.0],
    [2.22, 6.25],
    [6.11, 4.17]
]

matrix = euclidean_distances(X, X)
print(matrix)
with open("dist-matrix.csv", 'w') as fx:
    for row in matrix:
        for item in row:
            fx.write(str(item) + ",")
        fx.write("\n")
```
